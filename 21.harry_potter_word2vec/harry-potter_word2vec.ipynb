```python
import pandas as pd
import numpy as np
import gensim
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for filename in os.listdir('data'):
    
    f = open(os.path.join('data',filename), encoding='utf-8')
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,    #The maximum distance between the current and predicted word within a sentence
    min_count=2,    #Minimum word count threshold for inclusion into the word2vec model
    sg=1        #0 for CBOW and 1 for Skip-Gram
)
```

```python
model.build_vocab(story)     #build vocabulary
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)    #train the model
```
