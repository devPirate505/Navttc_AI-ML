```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python

df = pd.DataFrame({"Area": [7000, 7500, 9000, 8500, 8000, 8750], "Price": [300000, 400000, 700000, 500000, 470000, 520000]})

df
```

```python
%matplotlib inline

plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area, df.Price)

plt.show()
```

```python
X = pd.DataFrame(df['Area'])
y = df['Price']
```

```python
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Equation of line is y = {slope:2f} + {intercept:2f} * size")
```

```python
%matplotlib inline

plt.scatter(X, y, color='blue', label='Traning Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)

print(f"Mean Absolute error: {mae:2f}")
```
